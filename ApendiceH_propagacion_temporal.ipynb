```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
muestras = 50
tiempo = 20 + 200*np.random.rand(muestras)
tiempo.sort()
observable = 500 + 100*np.random.rand(muestras) - 0.000345*np.power(tiempo, 3) + 0.08248*np.power(tiempo, 2) - 2.5*tiempo
```

```python
tiempo = np.round(tiempo,1)
observable = np.round(observable,1)
plt.scatter(tiempo, observable)
plt.show()
```
