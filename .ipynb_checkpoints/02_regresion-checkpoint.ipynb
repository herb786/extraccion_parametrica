```python
'''
Ahora veremos el caso de un probema donde la información observada se asume que depende de dos factores.
En este caso extraeremos dos parámetros además del intercepto.
'''
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cbook, cm
import json
```

```python
# La entrada y la salida del problema tienen esta representacion
entrada1 = [10.561, 10.5648, 10.8129, 10.8308, 10.8619, 11.2876, 11.3351, 11.3511, 11.3635, 11.5446, 11.798, 11.9179, 12.1211, 12.4297, 12.465, 12.6014, 12.8191, 13.8946, 13.9428, 14.3759, 14.4462, 14.4676, 14.5719, 14.6461, 15.074, 15.1474, 15.4055, 15.5871, 15.6087, 15.6692]
entrada2 = [4.5503, 4.5584, 4.6473, 4.7865, 4.8835, 5.0418, 5.0752, 5.544, 5.5953, 5.7951, 5.8119, 5.8946, 6.0683, 6.0854, 6.2641, 6.2729, 6.3394, 6.3701, 6.3723, 6.3967, 6.5713, 7.1008, 7.1629, 7.2326, 7.4659, 7.5702, 7.6755, 7.6875, 7.7577, 7.8896]
salida = [215.0949, 199.6789, 201.2365, 229.3307, 207.0644, 231.2523, 223.1318, 225.3349, 205.3914, 263.8461, 260.4213, 232.3156, 227.6388, 290.6031, 237.193, 255.8111, 258.3252, 299.9257, 297.9813, 277.3957, 328.7366, 300.4824, 283.6778, 300.2182, 372.789, 295.1586, 335.2655, 353.8882, 360.6019, 360.3692]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(entrada1, salida, color="green")
ax2.scatter(entrada2, salida, color="red")
plt.show()
```

```python
en1 = np.array(entrada1).reshape(30,1)
en2 = np.array(entrada2).reshape(30,1)
salida = np.array(salida)
entrada = np.hstack((en1, en2))
```

```python
# Cargamos las rutinas de regresión lineal
reg = LinearRegression()
# Alimentamos los datos de entrada y salida
ajuste = reg.fit(entrada, salida)
```

```python
# Veámos la calificación
ajuste.score(entrada, salida)
```

```python
# En este caso tenemos dos parámteros debido a los dos factores tomados en cuenta para el análisis
ajuste.coef_
```

```python
# Usando los parámetros extraídos podemos estimar cuales hubieran sido los datos de salida bajo condiciones ideales
salida_ideal = reg.predict(entrada)
json.dumps(salida_ideal.tolist())
```

```python
# Veámos cuán distante se encuentra los valores estimados de los valores registrados realmente
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(entrada1, salida, color="green", marker='^')
ax1.scatter(entrada1, salida_ideal, color="red")
ax2.scatter(entrada2, salida, color="green", marker='+')
ax2.scatter(entrada2, salida_ideal, color="red")
plt.show()
```

```python
# Preparemos un rejilla bidimensional para crear una represantación de la regresión
mesh_x = np.linspace(10, 15, 20)
mesh_y = np.linspace(4, 8, 20)
mesh_z = []
for y in mesh_y:
    z = reg.predict(np.hstack((mesh_x.reshape(20,1), y*np.ones((20,1)))))
    mesh_z.append(z.tolist())
mesh_x, mesh_y = np.meshgrid(mesh_x, mesh_y) 
mesh_z = np.array(mesh_z)
```

```python
# Veámos una represenación tridimensional de los valores estimados y los registrados.
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(entrada1, entrada2, salida, marker='^')
ax.scatter(entrada1, entrada2, salida_ideal, marker='o')
surf = ax.plot_surface(mesh_x, mesh_y, mesh_z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0)
ax.view_init(15, 80, 0)
plt.show()
```
