```python
'''
La información disponible no siempre esta clasificada pero muestra patrones que permiten delucidar una forma de particionar esta.
Heurísticamente uno dispone el número de particiones observando los patrones en la representación gráfica.
Cada una de estas particiones tiene un centroide circundado por la información recogida.
El objetivo entonces es encontrar estos centroides de forma que la distancia con la información circundante sea mínima.
En argot matemático buscamos minimizar el segundo momento o varianza.
El amglicismo correspondiente para el repartimiento por centroides es "K-means" donde K denota el número de particiones fijado por el investigador.
'''
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cbook, cm
```

```python
entrada1 = 20 + 50*np.sin(2*np.pi*np.random.rand(30))
entrada2 = 30 + 30*np.sin(2*np.pi*np.random.rand(30))
```

```python
# La entrada del problema de clasificación
entrada1 = [26.82, 11.77, 19.84, 22.98, 12.39, 6.47, -2.26, 7.99, 18.19, 22.02, 25.44, -0.7, 11.08, -4.46, 24.75, -0.0, 9.44, 20.68, 4.91, 5.44, 33.3, 33.48, 33.19, 52.84, 38.42, 27.03, 40.51, 41.37, 47.47, 33.89, 47.22, 39.96, 32.0, 31.96, 42.14, 39.48, 40.07, 36.89, 30.86, 28.46, 41.28, 53.04, 53.72, 48.2, 51.41, 50.09, 53.13, 50.68, 53.52, 46.32, 50.86, 43.9, 60.62, 48.55, 51.23, 46.6, 50.15, 46.64, 56.54, 50.27]
```

```python
entrada2 = [13.57, 16.16, 29.27, 22.29, 19.71, 0.77, 34.11, 3.07, 22.53, 24.7, 25.06, 6.93, 18.22, 10.74, 13.42, 8.03, 10.54, 11.98, 37.08, 24.06, 68.13, 66.67, 69.68, 52.5, 69.25, 57.82, 57.79, 71.73, 47.04, 67.81, 50.49, 54.03, 63.4, 68.52, 73.18, 52.85, 59.94, 59.05, 66.25, 68.91, 5.27, 15.52, 3.79, 16.08, -2.55, -0.64, 10.61, 9.32, 5.78, 4.28, 9.98, 8.68, 0.98, 8.65, 3.89, 7.78, 10.11, 5.96, 13.74, 9.6]
```

```python
# Representmeos estos valores gráficamente para visualizar el número de particiones
plt.scatter(entrada1, entrada2)
plt.show()
```

```python
en1 = np.array(entrada1).reshape(60,1)
en2 = np.array(entrada2).reshape(60,1)
entrada = np.hstack((en1, en2))
```

```python
# Iniciemos las rutinas de repartimiento
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(entrada)
```

```python
# La clasificación retorna 3 categorías para cada par de números
kmeans.labels_
```

```python
# Los centroide son
kmeans.cluster_centers_
```

```python
plt.scatter(entrada1, entrada2, c=kmeans.labels_)
plt.scatter(40, 35, c='red', marker='+')
plt.show()
```

```python
# Veámos cómo seria clasificado este par de números
kmeans.predict([[40, 35]])
```
