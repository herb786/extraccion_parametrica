```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
num = 20
entrada1 = 10.56 + 5.2*np.random.rand(num)
entrada2 = 4.37 + 3.6*np.random.rand(num)
entrada3 = entrada2 +  2*np.random.rand(num)
entrada1.sort()
entrada2.sort()
entrada3.sort()
salida = 15*entrada1 + np.random.rand(num)*entrada1*entrada2 + 1.34*np.random.rand(num) + 5.5*entrada2 + entrada3
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
ax1.scatter(entrada1, salida, color="green")
ax2.scatter(entrada2, salida, color="red")
ax3.scatter(entrada3, salida, color="red")
plt.show()
```

```python
entrada1 = np.round(entrada1,2)
entrada2 = np.round(entrada2,2)
entrada3 = np.round(entrada3,2)
en1 = np.array(entrada1).reshape(num,1)
en2 = np.array(entrada2).reshape(num,1)
en3 = np.array(entrada3).reshape(num,1)
entrada = np.hstack((en1, en2, en3))
salida = np.round(salida,2)
```
