```python
'''
En la maraña de información adquirida, encontramos variables correlacionadas que permite simplificar el análisis con menos factores.
En álgebra estos factores son las componentes de un vector lineal. Verbi gratia, el binomio de coordenadas en un plano.
Podemos conseguir esta síntesis o simplificación usando la descomposición de vectores con un cambio de base vectorial.
Por ejemplo, una circunsferencia es representado por la dupla (x, y) en una base cardinal pero en una base polar la dupla es (r, θ).
Dado que el radio es constante hemos logrado explicar el problema con sólo la variable angular.
En resumen el problema es encontrar los vectores de este cambio de base.
El tecnicismo albionés correspondiente es PCA  traducido literalmente "análisis por componentes principales" donde las componente principales es la base vectorial que buscamos.
'''
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# La entrada del problema toma en cuenta tres factores
entrada = [[10.74, 4.67, 5.21], [11.03, 4.86, 5.59], [11.77, 5.01, 6.27], [11.92, 5.1, 6.33], [12.29, 5.31, 6.34], [12.34, 5.51, 6.67], [12.4, 5.55, 6.77], [12.63, 5.75, 6.95], [13.07, 6.14, 7.2], [13.1, 6.55, 7.22], [13.31, 6.61, 7.26], [13.37, 6.72, 7.42], [13.72, 6.75, 7.59], [13.77, 6.88, 7.78], [14.26, 7.13, 7.81], [14.77, 7.46, 8.01], [14.87, 7.59, 8.32], [15.03, 7.68, 8.39], [15.22, 7.7, 8.5], [15.74, 7.88, 9.29]]
```

```python
# La salida del problema
salida = [207.91, 205.14, 221.02, 248.7, 282.34, 249.94, 272.21, 238.48, 312.13, 273.43, 322.67, 283.71, 273.71, 341.29, 325.8, 270.97, 378.48, 364.26, 309.17, 324.71]
```

```python
entrada = np.array(entrada)
e1 = entrada.transpose()[0]
e2 = entrada.transpose()[1]
e3 = entrada.transpose()[2]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
ax1.scatter(e1, salida, color="green")
ax2.scatter(e2, salida, color="red")
ax3.scatter(e3, salida, color="blue")
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
ax1.plot(e1,e2)
ax2.plot(e1,e3)
ax3.plot(e2,e3)
plt.show()
```

```python
# Iniciamos las rutinas de síntesis para usar 2 factoes en lugar de 3
pca = PCA(n_components=2)
```

```python
# Verificamos el número de muestras
pca.n_samples_
```

```python
# Verificamos el número de factores
pca.n_features_in_
```

```python
# Realizamos el ajuste con los datos de entrada
pca.fit(entrada)
```

```python
# El trasvase retorna muestras dependientes de dos factores
# La salida no debe ser transformada porque se encuentra en otro espacio vectorial
pca.transform(entrada)
```

```python
pca_e = pca.transform(entrada)
e1 = pca_e.transpose()[0]
e2 = pca_e.transpose()[1]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharey=True)
ax1.scatter(e1, salida, color="green")
ax2.scatter(e2, salida, color="red")
plt.show()
```
