```python
'''
En los casos anteriores hemos buscados patrones desde el punto de vista geómetrico.
Una alternativa es buscar modelos estadísticos siguiendo las observaciones empíricas acumuladas.
Bayes postula que la probabilidad de nuestras observaciones cambiará según la información a priori. Es lo que se conoce como probabilidad condicional.
Ahora asumiendo que las observaciones siguen una determinada distribución de probabilidad buscaremos los parámetros que caracterizan tal distribución de probabilidad.
'''
```

```python
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
from matplotlib import cbook, cm
import numpy as np
```

```python
# Estudiemos un supuesta espora extrasolar y la sometemos a variaciones de temperatura y presión
experimento = [[81.0, 81.0], [74.0, 30.0], [90.0, 27.0], [48.0, 44.0], [23.0, 22.0], [17.0, 36.0], [53.0, 97.0], [33.0, 37.0], [93.0, 50.0], [34.0, 6.0], [26.0, 55.0], [99.0, 5.0], [99.0, 50.0], [84.0, 62.0], [7.0, 66.0], [29.0, 33.0], [45.0, 80.0], [72.0, 48.0], [81.0, 98.0], [52.0, 41.0], [100.0, 37.0], [88.0, 29.0], [36.0, 22.0], [36.0, 75.0], [97.0, 25.0], [98.0, 17.0], [68.0, 12.0], [92.0, 55.0], [37.0, 26.0], [3.0, 4.0], [96.0, 84.0], [45.0, 93.0], [54.0, 75.0], [31.0, 22.0], [29.0, 80.0], [98.0, 99.0], [18.0, 96.0], [61.0, 13.0], [71.0, 61.0], [99.0, 69.0]]
```

```python
# El supuesto estudio detectó un cabio en la coloración 
# Blanco identificado con el cardinal 0
# Verde identificado con el cardinal 1
# Naranja identificado con el cardinal 2
# Violeta identificado con el cardinal 3
salida = [0, 1, 1, 2, 3, 2, 2, 2, 1, 3, 2, 1, 1, 0, 2, 2, 2, 1, 0, 2, 1, 1, 3, 2, 1, 1, 1, 0, 3, 3, 0, 2, 2, 3, 2, 0, 2, 1, 0, 0]
```

```python
# Observemos la distribución de estas medidad en un plano cartesiano
entrada = np.array(experimento)
e1 = entrada.transpose()[0]
e2 = entrada.transpose()[1]
cs = plt.scatter(e1, e2, c=salida, cmap=cm.jet)
plt.colorbar(cs)
plt.show()
```

```python
# En este caso asumiremos una distribución normal e inicializamos la rutina correspondiente
clf = GaussianNB()
```

```python
# Calculamos los parámetros del modelo
clf.fit(entrada, salida)
```

```python
clf.predict([[60, 60]])
```
