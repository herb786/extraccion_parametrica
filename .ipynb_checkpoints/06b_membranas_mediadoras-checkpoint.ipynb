```python
from sklearn import svm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Veámos esta serie temporal siendo la entrada el tiempo transcurrido y la salida alguna propiedad que buscamos predecir
entrada = [1.1, 11.97, 16.75, 17.53, 18.84, 20.07, 22.86, 23.97, 25.26, 26.44, 28.6, 35.09, 38.22, 43.02, 45.9, 54.94, 56.5, 57.19, 57.45, 57.75, 61.67, 65.49, 70.7, 75.34, 78.94, 79.06, 79.09, 84.15, 86.0, 93.39]
```

```python
salida = [45.99, -41.61, 47.06, -27.14, -4.99, 31.84, 43.77, -40.59, 14.03, 26.06, 29.31, 41.42, 37.24, -52.88, -39.68, -7.75, -11.35, 38.88, 34.61, 24.62, -45.04, -52.21, -16.9, -19.8, 19.38, 30.27, 31.37, -61.99, 19.9, -71.55]
```

```python
entrada = np.array(entrada).reshape(30,1)
plt.scatter(entrada, salida)
plt.show()
```

```python
regr = svm.SVR(kernel="rbf", C=100, gamma=0.1, epsilon=0.1)
```

```python
regr.fit(entrada, salida)
```

```python
salida_pred = regr.predict(entrada)
```

```python
plt.scatter(entrada, salida)
plt.plot(entrada, salida_pred)
plt.show()
```
